## Review $\mathbb{L}_1$ vs $\mathbb{L}_2$ approximation

```python
from __future__ import divisionimport scipy.signal
```

Basic least squares problemmin $||\mathbf{x}||_2^2$where $x_1 + 2 x_2 = 1$

```python
from matplotlib.patches import Circlex1 = linspace(-1,1,10)fig=figure()ax=fig.add_subplot(111)ax.plot(x1,(1-x1)/2)ax.add_patch(Circle((0,0),1/sqrt(5),alpha=0.3))ax.plot(1/5,2/5,'rs')ax.axis('equal')ax.set_xlabel('$x_1$',fontsize=24)ax.set_ylabel('$x_2$',fontsize=24)ax.grid()
```

Now, let's do same problem in $L_1$ normmin $||\mathbf{x}||_1=|x_1|+|x_2|$where $x_1 + 2 x_2 = 1$In this case the constant-norm contour in the $L_1$ norm is a diamond-shape instead of a circle.

```python
from matplotlib.patches import Rectangleimport matplotlib.patchesimport matplotlib.transformsr=matplotlib.patches.RegularPolygon((0,0),4,1/2,pi/2,alpha=0.5)fig=figure()ax=fig.add_subplot(111)ax.plot(x1,(1-x1)/2)ax.plot(0,1/2,'rs')ax.add_patch(r)ax.grid()ax.set_xlabel('$x_1$',fontsize=24)ax.set_ylabel('$x_2$',fontsize=24)ax.axis('equal')
```
