The following is a classic estimation problem called the "lighthouse problem". The figure shows a set of receivers distributed at ccoordinates $x_k$ along the shore and a lighthouse located at some position $(\alpha,\beta)$ offshore. For this analysis, we assume that, $\beta=1$, which leaves $\alpha$ as unknown. The coastline is equipped with a continuous strip of photodetectors. The lighthouse flashes the shore $N$ times. The strip of  photodetectors registers the $k^{th}$ flash position $x_k$, but the the angle $\theta_k$ of the flash is unknown. Furthermore, the lighthouse beam is laser-like so there is no smearing along the strip of photodetectors. The problem is how to estimate $ \alpha  $.

From basic trigonometry, we have the following:

$$ \beta \tan(\theta_k) = x_k - \alpha $$

The density function for the angle is assumed to be the following:

$$ f_{\alpha}(\theta_k) = \frac{1}{\pi} $$

This means that the density of the angle is uniformly distributed between $ \pm \pi/2 $. Now, what we really want is the density function for $x_k$ which will tell us the probability that the $k^{th}$ flash will be recorded at position $ x_k $. After a transformation of variables, we obtain the following:

$$ f_{\alpha}(x_k) = \frac{\beta}{\pi(\beta ^2 +(x_k-\alpha)^2)} $$

which we plot below for some reasonable factors


```python
xi = linspace(-10,10,150)
alpha = 1
f = lambda x: 1/(pi*(1+(x-alpha)**2))
plot(xi,f(xi))
xlabel('$x$',fontsize=24)
ylabel(r'$f_{\alpha}(x)$',fontsize=24);
vlines(alpha,0,.35)
grid()
```

As shown in figure, the peak for this distribution is when $ x_k=\alpha $. Because there is no coherent processing, the recording of a signal at one position does not influence what we can infer about the position of another measurement. Thus, we can justify assuming independence between the individual $x_k$ measurements. Then, given $N$ measurements, we form the likelihood as the following:

$$ L(\alpha) = \prod_{k=1}^N f_{\alpha}(x_k)$$

And this is the function we want to maximize for our maximum likelihood estimator. Let's process the logarithm of this to  make the product easy to work with as this does not influence the position of the maximum $ \alpha $. Then,

$$\mathcal{L}(\alpha)=\sum_{k=1}^N \ln f_{\alpha}(x_k) = \texttt{constant}- \sum_{k=1}^N \ln (\beta ^2 +(x_k-\alpha)^2) $$

Taking the first derivative gives us the equation would have to solve in order to compute the estimator for $ \alpha $,

$$ \frac{d \mathcal{L}}{d \alpha} = 2 \sum_{k=1}^N \frac{x_k-\alpha}{\beta^2+(x_k-\alpha)^2}=0$$

Unfortunately, there is no easy way to solve for the optimal $ \alpha $ for this equation.

```python
beta  =alpha = 1
theta_samples=((2*np.random.rand(250)-1)*pi/2)
x_samples = alpha+beta*np.tan(theta_samples)
hist(x_samples);
```

```python
import sympy as S
a=S.symbols('alpha')
#S.init_printing()
```

```python
L=sum((xk-a)/(1+(xk-a)**2)  for xk in x_samples)
```

```python
S.plot(L,(a,-15,15))
```

There are tools in `scipy.optimize` that can help us compute the zero-crossing as demonstrated in the cell below.

```python
from scipy import optimize
from scipy.optimize import fmin_l_bfgs_b

alpha_x = optimize.fsolve(S.lambdify(a,L),0)
alpha_x  = fmin_l_bfgs_b(S.lambdify(a,(L)**2),0,bounds=[(-3,3)],approx_grad=True)

print alpha_x
```

However, to avoid this problem, we could just look at the log likelihood function and observe that it is concentrated around the mean. Then, we could just use the usual average of the $x_k$ as our estimator instead of having to solve that complicated equation.

```python
x = S.symbols('x')
tmp=sum(S.log(1/(pi*(1+(x-a)**2))).subs(x,i) for i in x_samples[:20])
S.plot(tmp,(a,-15,15),xlabel=r'$\alpha$');
```

```python
def run_trials(n=100):
    o=[]
    for i in range(100):
        theta_samples=((2*np.random.rand(250)-1)*pi/2)
        x_samples = alpha+beta*np.tan(theta_samples)
        o.append(x_samples)
    return np.array(o)
```

## Maximum A-Posteriori (MAP) Estimation

Let's use a uniform distribution for the prior of $\alpha$ around some bracketed interval

$$ f(\alpha) = \frac{1}{\alpha_{max}-\alpha_{min}} \quad where \quad \alpha_{max} \le \alpha \le \alpha_{min}$$

and zero otherwise. We can compute this sample-by-sample to see how this works using this prior.

```python
alpha  = a
alphamx,alphamn=-3,3
g = f(x_samples[0])
xi = linspace(alphamn,alphamx,100)
mxval = S.lambdify(a,g)(xi).max()
plot(xi,S.lambdify(a,g)(xi),x_samples[0],mxval*1.1,'o')
```

```python
xi = linspace(alphamn,alphamx,100)
def slide_figure(n=1):
    fig,ax=subplots()
    g = S.prod(map(f,x_samples[:n]))
    mxval = S.lambdify(a,g)(xi).max()*1.1
    ax.plot(xi,S.lambdify(a,g)(xi))
    ax.plot(x_samples[:n-1],mxval*ones(n-1),'o',color='gray',alpha=0.3)
    ax.plot(x_samples[n],mxval,'o',ms=10.)
    ax.set_xlabel(r'$\alpha$',fontsize=18)
    ax.axis(ymax=mxval*1.1)
```

```python
interact(slide_figure, n=(1,15,1));
```
