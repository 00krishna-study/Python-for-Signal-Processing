### FIR Filters

We are primarily interested in filters of the following type:

$$ y_n = \sum_{k=0}^{M-1} h_n x_{n-k} $$

with real input $ x_n $ and real output $ y_n $. These are called *finite* impulse response because they stop after running out of input (i.e. there is no feedback to keep this going indefinitely). These are also sometimes called *moving average* filters. For example, given $ h_0 = h_1 = 1/2 $, we have

$$ y_n = x_{n}/2 + x_{n-1}/2 $$

Thus, if the input $ x_n=1 \hspace{.5em} \forall \hspace{.5em} n \ge 0 $, then $ y_n = 1 \hspace{.5em} \forall \hspace{.5em} n > 1  $. Note that we lose one sample in filling the filter for $ n=0 $. In other words, we have to wait one sample for a valid filter output. On the other hand, if 
 $ x_n= \exp \left( j\pi n \right)  \hspace{.5em} \forall \hspace{.5em} n \ge 0 $, then $ y_n=0  $. This means that this moving average has filtered out the highest frequency signal ($ \Omega=\pi $) and preserved the lowest frequency signal ($ \Omega=0 $)

```python
from __future__ import division
from scipy import signal 

def db20(W,Nfft=None):
    'Given DFT, return power level in dB or take DFT if need be'
    assert  np.isscalar(W) or W.ndim==1 
    if Nfft is None: # assume W is DFT 
        return 20*log10(abs(W))
    else: # assume time-domain passed, so need DFT
        DFT= fft.fft(array(W).flatten(),Nfft)/sqrt(Nfft)
        return 20*log10(abs(DFT.flatten()))

def undb20(x):
    'Invert the db20. Recover amplitude'
    return 10**(x/20.)
```

```python
fig, axs = subplots(2,1,sharex=True)
subplots_adjust( hspace = .2 )
fig.set_size_inches((5,5))

ax=axs[0]
w,h=signal.freqz([1/2., 1/2.],1)
ax.plot(w,20*log10(abs(h)))
ax.set_ylabel(r"$ 10 \log_{10} |H(\omega)| $",fontsize=18)
ax.grid()

ax=axs[1]
ax.plot(w,angle(h)/pi*180)
ax.set_xlabel(r'$\omega$ (radians/s)',fontsize=18)
ax.set_ylabel(r"$\phi $ (deg)",fontsize=18)
ax.set_xlim(xmax = pi)
ax.grid()
```

The figure in the top plot shows the amplitude response of the filter and the figure below is the phase response in degrees. At $\omega=0 $, we have $|H(\omega)|=1 $ which says that our moving average filter does not change the amplitude of the signal. We observed this earlier with $x_n=1 $ that produced $y_n=1 $. When we consider the other extreme with $\omega= \pi $, we have $|H(\omega)|=0$ which we observed earlier for $ x_n= \exp \left( j\pi n \right)  \hspace{.5em} \forall \hspace{.5em} n \ge 0 $.

Now, let's consider a signal 

$$ x_n= \exp \left( j\pi n/2 \right)  \hspace{.5em} \forall \hspace{.5em} n \ge 0 $$

which is halfway between these two extremes

```python
Ns=30
n= arange(Ns)
x = cos(arange(Ns)*pi/2.)
y= signal.lfilter([1/2.,1/2.],1,x)

fig,ax = subplots(1,1)
fig.set_size_inches(9,4)

ax.stem(n,x,label='input',basefmt='b-')
ax.plot(n,x,':')
ax.stem(n[1:],y[:-1],markerfmt='ro',linefmt='r-',label='output')
ax.plot(n[1:],y[:-1],'r:')
ax.set_xlim(xmin=-1.1)
ax.set_ylim(ymin=-1.1,ymax=1.1)
ax.set_xlabel("sample",fontsize=18)
ax.legend(loc=0)
ax.set_ylabel("amplitude",fontsize=18);
```

The figure above shows the input/output response of the filter when  $\omega=\pi/2 $. At this point, $ |H(\omega)|^2 = 1/2 $ meaning the signal energy has been cut in half. The signal phase has been shifted by 45 degrees. To see this, note that the input signal repeats every four samples (360 degrees). Thus a shift of 45 degrees is equivalent to a shift of one sample. Note that `signal.lfilter` automatically inserts a zero initial condition so I had to drop that one incomplete sample to line up the plots.
