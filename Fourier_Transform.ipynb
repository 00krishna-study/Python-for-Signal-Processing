```python
from __future__ import  division
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrow
import mpl_toolkits.mplot3d.art3d as art3d
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def dftmatrix(Nfft=32,N=None):
    k= np.arange (Nfft)
    if N is None: N = Nfft
    n = arange (N)
    U = matrix(exp(1j* 2*pi/Nfft *k*n[:,None]))
    return U/sqrt(Nfft)

Nfft=16
v = ones((16,1))
U = dftmatrix(Nfft=Nfft,N=16)

f,ax=subplots()
ax.set_aspect(1)
a=2*pi/Nfft*arange(Nfft)
ax.plot( cos(a),sin(a), '.')
for i in a:
  ax.add_patch( FancyArrow(0,0,cos(i),sin(i),width=0.02,length_includes_head=True))

ax.text(1,0,'0',fontsize=16)
ax.text(0.1,1,r'$\frac{\pi}{2}$',fontsize=22)
ax.text(-1,0.1,r'$\pi$',fontsize=18)
ax.text(0.1,-1,r'$\frac{3\pi}{2}$',fontsize=22)
ax.axis(array([-1,1,-1,1])*1.25)
plt.show()

```

```python


fig = plt.figure()
fig.set_size_inches(6,6)
ax = fig.add_subplot(1, 1, 1, projection='3d')
ax.axis([-1,1,-1,1])
ax.set_zlim([0,d[:,2].max()])
ax.set_aspect(1)
ax.view_init(azim=-30)
ax.set_xlabel('real')
ax.set_ylabel('imag')
ax.set_zlabel('Abs')

def facet_filled(x,alpha=0.5,color='b'):
    'construct 3D facet from adjacent points filled to zero'
    a,b=x
    a0= a*array([1,1,0])
    b0= b*array([1,1,0])
    ve = vstack([a,a0,b0,b])      # create closed polygon facet
    poly = Poly3DCollection([ve]) # create facet
    poly.set_alpha(alpha)
    poly.set_color(color)
    return poly

sl=[slice(i,i+2) for i in range(d.shape[0]-2)] # collect neighboring points
for s in sl:
  poly=facet_filled(d[s,:])
  ax.add_collection3d(poly)
 
# edge polygons    
ax.add_collection3d(facet_filled(d[[-1,0],:]))
ax.add_collection3d(facet_filled(d[[-2,-1],:]))


plt.show()
```
