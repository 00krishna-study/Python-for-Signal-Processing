```python
from matplotlib.lines import Line2D
from matplotlib.patches import Rectangle
```

```python
def needle(ang=0,L=0.5):
    xc,yc = rand(2)
    x0,y0 = xc-L*cos(ang),yc-L*sin(ang)
    xe,ye = xc+L*cos(ang),yc+L*sin(ang)
    if (0<x0<1) and (0<y0<1) and (0<xe<1) and (0<ye<1):
        needle.blue_count+=1
        return ([Line2D([x0,xe],[y0,ye],alpha=0.3) , Line2D([xc],[yc],marker='o',alpha=0.3)] )
    else:
        needle.red_count+=1
        return ([Line2D([x0,xe],[y0,ye],color='r',alpha=0.3) , Line2D([xc],[yc],marker='o',color='r')] )
```

```python
fig,ax=subplots()
needle.red_count=0 # reset counters
needle.blue_count=0
for k in range(500):
    for i in needle(rand()*pi,.3): ax.add_line(i)

ax.set_aspect(1)
ax.add_patch(Rectangle((0,0),1,1,alpha=0.3,color='y'))
ax.axis([-.1,1.1,-.1,1.1])
ax.set_title('Prob inside square=%3.3f'% (needle.blue_count/(needle.blue_count+needle.red_count)))
ax.grid()
```
