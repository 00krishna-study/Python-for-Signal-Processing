In the prior section, we discovered that resolving two nearby frequencies depends crucially on the rectangular window and its corresponding `sinc` Discrete Fourier Transform (DFT). However, our two nearby signals had *equal* amplitude. What happens if this is no longer the case? The figure below shows the DFT of two nearby signals where one has ten times the amplitude. 

```python
from __future__ import  division
from scipy import signal

# some useful functions 
def dftmatrix(Nfft=32,N=None):
    'construct DFT matrix'
    k= np.arange(Nfft)
    if N is None: N = Nfft
    n = arange(N)
    U = matrix(exp(1j* 2*pi/Nfft *k*n[:,None])) # use numpy broadcasting to create matrix
    return U/sqrt(Nfft)


def db20(W,Nfft=None):
    'Given DFT, return power level in dB'
    if Nfft is None: # assume W is DFT 
        return 20*log10(abs(W))
    else: # assume time-domain passed, so need DFT
        return 20*log10(abs( fft.fft(W,Nfft)/sqrt(Nfft) ) )

fs = 64 # sampling frequency
t = arange(0,2,1/fs)
f = 10  # one signal
deltaf = 2.3 # second nearby frequency

Nf = 512
fig,ax = subplots(2,1,sharex=True)
fig.set_size_inches((9,4))

x=cos(2*pi*f*t) + cos(2*pi*(f+deltaf)*t) # equal amplitudes
X = fft.fft(x,Nf)/sqrt(Nf)
ax[0].plot(linspace(0,fs,len(X)),abs(X),'-o',ms=3.)
ax[0].set_title(r'Nearby Frequencies of Different Amplitude Not Resolvable',fontsize=14)
ax[0].set_ylabel(r'$|X(k)|$',fontsize=18)
ax[0].set_xlim(xmax = fs/2)
ax[0].grid()
ax[0].text(0.5,0.5,'Same amplitude',
            transform=ax[0].transAxes,
            backgroundcolor='Lightyellow',
            fontsize=16)

x=cos(2*pi*f*t) + 10*cos(2*pi*(f+deltaf)*t) # one has 10x the amplitude
X = fft.fft(x,Nf)/sqrt(Nf)
ax[1].plot(linspace(0,fs,len(X)),abs(X),'-o',ms=3.)
ax[1].set_ylabel(r'$|X(k)|$',fontsize=18)
ax[1].set_xlabel('Frequency (Hz)',fontsize=18)
ax[1].set_xlim(xmax = fs/2)
ax[1].grid()
ax[1].text(0.5,0.5,'Different amplitude',
                transform=ax[1].transAxes,
                backgroundcolor='lightyellow',
                fontsize=16)
ax[1].annotate('Smaller signal',
            fontsize=12,xy=(f,abs(X)[int(f/fs*Nf)]),
            xytext=(2,15),
            arrowprops={'facecolor':'m','alpha':.3})
```

In the figure above, the top plot shows the  two peaks of  nearby signals when they both have the same amplitude. The bottom plot shows the case where the higher frequency signal has ten times the amplitude. Note that the two frequencies are no longer resolvable because the weaker signal (smaller amplitude) has descended into the sidelobes of the  stronger one. Thus, we need more than just longer sampling durations and denser DFTs to resolve these signals. This is where *windows* (sometimes called *tapers*) come in. 

### What is a window?

A window is a function that is multiplied sample-by-sample against the input signal (i.e. element-wise).  For example, given an input signal, $s_n$, and a window, $w_n$, the new windowed signal is,

$$ y_n = s_n w_n $$

known in vector notation as the *Hadamard* product,

$$ \mathbf{y} = \mathbf{w} \odot \mathbf{s}  $$

The following figure compares the the rectangular and triangular windows applied to an input sinusoid.

```python
fig,axs = subplots(2,1)
fig.set_size_inches((10,5))

Nf = 128
nsamp = 64
window = signal.triang(nsamp)
rectwin = ones(nsamp)
ax = axs[0]

ax.plot(arange(nsamp),window,'m-o',label='triangular')
ax.stem(arange(nsamp),rectwin,basefmt='b-',label='rectangular')
ax.set_ylim(ymax = 1.1,ymin=-0.1)
#ax.set_xlim(xmin=-1,xmax = 17)
ax.legend(loc=0)

x=array(dftmatrix(64,64)[:,5].real).flatten() # convert to numpy array
n = arange(len(x))
window = signal.triang(len(x))

ax = axs[1]
ax.plot(n,x,'-o',label='signal',ms=3.,alpha=0.3)
ax.plot(n,window*x,'-or',label='windowed\nsignal',ms=5.)
ax.set_ylabel('Amplitude',fontsize=16)
ax.set_xlabel('sample index',fontsize=16)
ax.legend(loc=0)
ax.grid()
ax2 = ax.twinx()
ax2.fill_between(n,window,alpha=0.2,color='m')
ax2.set_ylabel('Window function',fontsize=16)
```

In the figure above, the top plot shows the $w_n$ sequences for the <font color="blue"> rectangular</font> and <font color="magenta"> triangular</font> windows. The bottom plot shows the input <font color="blue">signal before (light blue line) </font> and <font color="red"> after (red line)</font> applying the <font color="magenta">triangular window</font>. The amplitudes are noted on the vertical scale on the left. The triangular window is drawn in the background with scale noted on the right side. Basicially, the red line is the product of the blue line and the pink background window function. 

As shown, because the triangular window slopes to zero at  the beginning and at the end, the amplitude of the input signal drops away from the middle. This means that we lose precious signal energy at the start/end of the sequence and we certainly want something in exchange. To examine the  trade-offs, we need a framework to evaluate different window functions. 

The following figure shows the two input signals with rectangular, triangular, and Hamming windows applied.

```python
Nf = 512
fig,ax = subplots(3,1,sharex=True,sharey=True)
fig.set_size_inches((9,4))

x=cos(2*pi*f*t) + 10*cos(2*pi*(f+deltaf)*t)
X = fft.fft(x,Nf)/sqrt(Nf)
ax[0].plot(linspace(0,fs,len(X)),db20(X),'-o',ms=3.)
#ax[0].set_title(r'Nearby Frequencies of Different Amplitude Not Resolvable',fontsize=14)
ax[0].set_ylabel(r'$|X(k)|$',fontsize=18)
ax[0].set_xlim(xmax = fs/2)
ax[0].grid()
ax[0].text(0.5,0.5,'rectangular window',
                transform=ax[0].transAxes,
                backgroundcolor='lightyellow',
                fontsize=16)
ax[0].annotate('Smaller signal',
            fontsize=12,xy=(f,db20(X)[int(f/fs*Nf)]),
            xytext=(1,30),
            arrowprops={'facecolor':'m'})

w = signal.triang(len(x))
X = fft.fft(x*w,Nf)/sqrt(Nf)
ax[1].plot(linspace(0,fs,len(X)),db20(X),'-o',ms=3.)
ax[1].set_ylabel(r'$|X(k)|$',fontsize=18)
ax[1].set_xlim(xmax = fs/2)
ax[1].grid()
ax[1].text(0.5,0.5,'triangular window',
                transform=ax[1].transAxes,
                backgroundcolor='lightyellow',
                fontsize=16)
ax[1].annotate('Smaller signal',
            fontsize=12,xy=(f,db20(X)[int(f/fs*Nf)]),
            xytext=(1,30),
            arrowprops={'facecolor':'m'})

w = signal.hamming(len(x))
X = fft.fft(x*w,Nf)/sqrt(Nf)
ax[2].plot(linspace(0,fs,len(X)),db20(X),'-o',ms=3.)
ax[2].set_ylabel(r'$|X(k)|$',fontsize=18)
ax[2].set_xlabel('Frequency (Hz)',fontsize=18)
ax[2].set_xlim(xmax = fs/2)
ax[2].grid()
ax[2].set_ylim(ymin=-20)
ax[2].text(0.5,0.5,'Hamming window',
                transform=ax[2].transAxes,
                backgroundcolor='lightyellow',
                fontsize=16)
ax[2].annotate('Smaller signal',
            fontsize=12,xy=(f,db20(X)[int(f/fs*Nf)]),
            xytext=(1,30),
            arrowprops={'facecolor':'m'})
```

To enhance detail, the figure above shows the the $20\log_{10} | X_k |$ of the DFT of the input signal with various windows applied on the same scale. The top plot shows the rectangular window and it is very difficult to make out the smaller signal in the sidelobes of the dominant larger signal. The middle plot shows the triangular window and we can see that there is definitely something peeking out from the sidelobes, but we can see this most clearly using the Hamming window, which shows a peak at the nearby frequency. Note that the width of the mainlobe widened and as the second weaker signal emerged from the sidelobes of the dominant larger amplitude signal. This is one of the trade-offs involved with using windows for spectral analysis.

### Signal-to-Noise Ratio

To analyze the effect of windows, we define the signal-to-noise ratio and then compute it before and after applying the window function. The input signal power is,

$$ ||\mathbf{s}||^2 = |\sum_{n=0}^{N_s} s_n|^2 = |\mathbf{1}^T \mathbf{s}|^2$$

The input noise power is

$$ ||\boldsymbol{\nu}||^2 = \mathbb{E}\left( \sum_{n=0}^{N_s} \nu_n \right)^2 = N_s \sigma_\nu^2$$

where the noise power is zero-mean and mutually-uncorrelated. Thus, the input signal-to-noise is

$$SNR_{in} = \frac{|\mathbf{1}^T \mathbf{s}|^2}{N_s \sigma_\nu^2}$$

Likewise, after windowing, the signal-to-noise ratio is

$$SNR_{out} = \frac{|\mathbf{w}^T \mathbf{s}|^2}{ \mathbf{w}^T \mathbf{w} \sigma_\nu^2}$$

Finally, the ratio of the before and after signal-to-noise ratios is defined as the *processing gain*,

$$ G_p = \frac{SNR_{out}}{SNR_{in}} = \frac{|\mathbf{w}^T \mathbf{s}|^2}{ \mathbf{w}^T \mathbf{w} \sigma_\nu^2}  \frac{N_s \sigma_\nu^2}{|\mathbf{1}^T \mathbf{s}|^2}$$

We take the input signal at one of the DFT bin frequencies  (i.e. $\Omega_k = \frac{2\pi}{N}$) to avoid the *scalloping* of the DFT in between bin frequencies. Without loss of generality, we can choose $\mathbf{s}=\mathbf{1}$ and then simplify the processing gain to obtain,

$$ G_p \triangleq \frac{|\mathbf{1}^T \mathbf{w}|^2}{\mathbf{w}^T\mathbf{w}}$$

The *coherent gain* is defined the ratio of the *amplitudes* of the input and output signals,

$$ G_{coh} \triangleq \mathbf{1}^T \mathbf{w}$$

### Noise Equivalent Bandwidth

The next metric we need to understand windows is the *Noise Equivalent Bandwidth*. Consider the DFT of the window function, $W_k$. The idea of Noise Equivalent Bandwidth is to  apply the window to a pure noise signal and compare this to what an ideal bandlimited filter would get.

$$\DeclareMathOperator{\Tr}{Trace}
\DeclareMathOperator{\bnu}{\boldsymbol{\nu}}
\mathbb{E} || \mathbf{w}^T\bnu ||^2 =  \Tr \mathbf{w}^T \mathbb{E} \left(\bnu \boldsymbol{\nu}^T \right) \mathbf{w} = \sigma_{\nu}^2 \mathbf{w}^T \mathbf{w }$$

We want to equate this power to the corresponding output noise power of a perfect bandlimited filter with height equal to $W_0$ and width $2 B_{neq}$. 

$$  2 B_{neq} \cdot W_0^2 \sigma_\nu^2 = \sigma_{\nu}^2 \mathbf{w}^T \mathbf{w } $$

and solving for $B_{neq}$ gives,

$$  B_{neq}  = \mathbf{w}^T \mathbf{w } / W_0^2 /2  $$

and since $W_0 = \mathbf{1}^T \mathbf{ w}$, we can write this as the following,

$$  B_{neq}  \triangleq \frac{1}{2}\frac{\mathbf{w}^T \mathbf{w } }{ |\mathbf{1}^T \mathbf{w}|^2 } = \frac{1}{2} \frac{1}{G_p} $$

### 6-dB Bandwidth

The 6-dB bandwidth measures how far down the main lobe of the DFT represents the half-power level. This is because $20 \log_{10}(1/2) \approx -6 $. The following figure shows the DFT for the  rectangular  window and its corresponding half-power point down the mainlobe. In general,  there is no closed form solution to the half-power level so you have to compute it numerically.

### Peak Sidelobe Level

```python
from scipy.interpolate import UnivariateSpline

fig,ax = subplots()

def sidelobe_level(W,return_loc=False):
    Nf = len(W)
    i,= where(abs(W)==0) # edges of lobes
    freqs = arange(Nf)/Nf*2. # normalized to Nyquist rate
    ix=where(np.logical_and(i[0]/Nf*2 < freqs , freqs< i[1]/Nf*2))[0]
    s=UnivariateSpline(ix/Nf*2, db20(W[ix]))
    ns = lambda x: -s(x)
    val= scipy.optimize.fmin( ns, mean(x),disp=0)[0]
    if return_loc:
        return (db20(W[0]) - s(val),val)
    else:
        return db20(W[0]) - s(val)
    
def db10(W):
    return 10*log10(abs(W))

def db20(W):
    return 20*log10(abs(W))

Nf = 256
w = ones(16)
W = fft.fft(w,Nf)/sqrt(Nf)

# find half-power level using scipy.optimize
freqs = arange(Nf)/Nf*2. # normalized to Nyquist rate

ax.plot(freqs ,db20(W))
w = signal.triang(16)
W = fft.fft(w,Nf)/sqrt(Nf)
ax.plot(freqs,  db20(W/W[0]))
ax.set_xlim(xmax = 0.5)
#ax.set_ylim(ymin=-10)
ax.grid()
```

The key to understanding windowing is to understand their DFTs. The figure below shows the DFTs of the rectangular window and the triangular window.

```python
from scipy import signal

fig,ax = subplots()

Nf = 128
window = signal.triang(16)
rectwin = ones(16)

ax.stem(arange(16),rectwin,basefmt='b-',label='rectangular')
ax.plot(arange(16),window,'r-o',label='triangular')
ax.set_ylim(ymax = 1.1,ymin=-0.1)
ax.set_xlim(xmin=-1,xmax = 17)
ax.legend(loc=0)
ax.set_xlabel('sample index',fontsize=16)

fig, ax = subplots(2,1,sharex=True)

freqs = arange(Nf)*2*pi/Nf
R = fft.fft(rectwin,Nf)/sqrt(Nf)
W = fft.fft(window,Nf)/sqrt(Nf)

ax[0].plot(freqs,abs(R),'-bo',ms=4.,label='rectangular')
ax[0].plot(freqs,abs(W),'-ro',ms=4.,label='triangular')
ax[0].set_ylabel(r'$|X|$',fontsize=14)
ax[0].legend(loc=0)

ax[1].plot(freqs,20*log10(abs(R)),'-bo',ms=4.)
ax[1].plot(freqs,20*log10(abs(W)),'-ro',ms=4.)
ax[1].set_ylim(ymin = -60,ymax=20)
ax[1].set_xlim(xmax = pi*1.01)
ax[1].set_xticks([0,pi/4.,pi/2.,3*pi/4.,pi])
ax[1].set_xticklabels(['0',r'$\frac{\pi}{4}$',r'$\frac{\pi}{2}$',r'$\frac{3\pi}{4}$',r'$\pi$' ],fontsize=18)
ax[1].set_xlabel('Discrete Frequency',fontsize=14)
ax[1].set_ylabel(r'$20\log_{10}(|X|)$',fontsize=14)
```

The figure above shows the magnitude of the respective DFTs for the rectangular and triangular windows. The middle plot shows the DFT magnitude in linear units on the bottom plot in in decibels (dB) (i.e. $20\log_{10}(|X|)$). From now on, we will focus on the dB plot because it reveals much more of the underlying structure of both DFTs.
