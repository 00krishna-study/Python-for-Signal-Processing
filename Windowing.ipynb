```python
def dftmatrix(Nfft=32,N=None):
    'construct DFT matrix'
    k= np.arange(Nfft)
    if N is None: N = Nfft
    n = arange(N)
    U = matrix(exp(1j* 2*pi/Nfft *k*n[:,None])) # use numpy broadcasting to create matrix
    return U/sqrt(Nfft)
```

```python
from scipy import signal

fig,ax = subplots()

Nf = 128
window = signal.hamming(16)
rectwin = ones(16)

ax.stem(arange(16),rectwin,basefmt='b-',label='rectangular')
ax.plot(arange(16),window,'r-o',label='hamming')
ax.set_ylim(ymax = 1.1,ymin=-0.1)
ax.set_xlim(xmin=-1,xmax = 17)
ax.legend(loc=0)
ax.set_xlabel('sample index',fontsize=16)

fig, ax = subplots(2,1,sharex=True)

freqs = arange(Nf)*2*pi/Nf
R = fft.fft(rectwin,Nf)/sqrt(Nf)
W = fft.fft(window,Nf)/sqrt(Nf)

ax[0].plot(freqs,abs(R),'-bo',ms=4.,label='rectangular')
ax[0].plot(freqs,abs(W),'-ro',ms=4.,label='hamming')
ax[0].legend(loc=0)

ax[1].plot(freqs,20*log10(abs(R)),'-bo',ms=4.)
ax[1].plot(freqs,20*log10(abs(W)),'-ro',ms=4.)
ax[1].set_ylim(ymin = -60,ymax=20)
ax[1].set_xlim(xmax = pi)
ax[1].set_xlabel('Discrete Frequency',fontsize=14)
```

The figure above shows the magnitude of the respective DFTs for the rectangular and hamming windows.
